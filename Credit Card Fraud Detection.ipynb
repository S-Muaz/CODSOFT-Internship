<h1> Credit Card Fraud Detection :</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ccdf=pd.read_csv("creditcard.csv")
```

```python
ccdf.head(5)
```

```python
ccdf.describe()
```

<h3>DATA PREPROCESSING :</h3>

```python
ccdf.isnull().sum()
```

```python
ccdf['Class'].value_counts()
```

<h4>We observe that there is an imbalance in our dataset hence we redefine our Dataframe :</h4>

```python
normal=ccdf[ccdf.Class==0]
```

```python
fraud=ccdf[ccdf.Class==1]
```

```python
print(normal.shape)
```

```python
print(fraud.shape)
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
ccdf.groupby('Class').mean()
```

```python
normal_sample=normal.sample(n=492)
```

<h4>Creating Balanced Dataset :</h4>

```python
new_file=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_file.head(10)
```

```python
new_file['Class'].value_counts()
```

```python
new_file.groupby('Class').mean()
```

```python
X=new_file.drop(columns='Class',axis=1)
```

```python
Y=new_file['Class']
```

<h4>Once DataFrame is ready we split it into training and testing</h4>
<h5>Train Data : 80% </h5>
<h5>Test Data :  20%</h5>

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression(max_iter=100)
```

```python
model.fit(X_train,y_train)
```

<h4>Making Predictions :</h4>

```python
X_train_prediction=model.predict(X_train)
```

```python
training_data_acuracy=accuracy_score(X_train_prediction,y_train)*100
```

<h4>Training Accuracy :</h4>

```python
print(f"Training Data Accuracy: {training_data_acuracy}%")
```

```python
X_test_prediction=model.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(X_test_prediction,y_test)*100
```

<h4>Testing Accuracy :</h4>

```python
print(f"Test Data Accuracy: {test_data_accuracy}%")
```
